```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("raw_dataset.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
mean_values = df.mean(numeric_only=True)
median_values = df.median(numeric_only=True)
variance_values = df.var(numeric_only=True)

mean_values, median_values, variance_values
```

```python
#visualise distributions 
## HISTOGRAM 
df.hist(figsize=(10,6))
plt.suptitle("Feature Distributions")
plt.show()
```

```python
#Shows relationships between features
## SCATTER PLOT 
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df)
plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
#For outlier detection
##BOX PLOTS
plt.figure(figsize=(8,6))
sns.boxplot(data=df.drop(columns=['species']))
plt.title("Boxplot of Numerical Features")
plt.show()
```

```python
#CORRELATION MATRIX
correlation_matrix = df.drop(columns=['species']).corr()
correlation_matrix
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

## EDA Insights

- Petal length and petal width show a strong positive correlation.
- Sepal features show moderate correlation.
- Different flower species are clearly separable based on petal measurements.
- No significant outliers were observed in the numerical features.

### Conclusion
Exploratory data analysis helped understand the structure and relationships in the dataset and provided valuable insights for future modeling.
